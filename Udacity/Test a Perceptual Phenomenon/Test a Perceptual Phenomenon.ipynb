# Test a Perceptual Phenomenon

## 1. Overview

For this project, we will test a classic psychology called Stroop effect:

>In psychology, the Stroop effect is a demonstration of interference in the reaction time of a task. When the name of a color (e.g., "blue", "green", or "red") is printed in a color that is not denoted by the name (e.g., the word "red" printed in blue ink instead of red ink), naming the color of the word takes longer and is more prone to errors than when the color of the ink matches the name of the color.

We will use data collected from an experiment which has two groups, congruent and incongruent,  each group recorded the time.
We then perform hypothesis test whether mean of congruent is equal to incongruent group.

## 2. Independent variable and dependent variable

Here we have word condition as independent variable , time is dependent variable

## 3. What hypothesis test should we use for this experiment?

In this experiment , we use t-test since it's size sample is less than 30, and population standard deviation is unknown.

Null Hypothesis:There is no significant response time difference between congruent group and incongruent group.
Alternative Hypothesis: There is a significant response time difference between congruent group and incongruent group.

## 4. Check with data

### 4.1 Load data

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
troop=pd.read_csv("Troop.csv")
```

```python
troop.head()
```

### 4.1 Check mean and standard deviation


```python
congruent_mean=troop["Congruent"].mean()
congruent_std=troop["Congruent"].std()
incongruent_mean=troop["Incongruent"].mean()
incongruent_std=troop["Incongruent"].std()
```

```python
mean_std=pd.DataFrame([{"Mean":congruent_mean,"Std":congruent_std},{"Mean":incongruent_mean,"Std":incongruent_std}],index=["Congruent","Incongruent"])
mean_std
```

### 4.2 Check distribution between Congruent and Incongruent

```python
plt.plot(troop)
plt.ylabel("Time")
plt.legend(["Congruent","Incongruent"])
plt.show()
```

Visually, we can see congruent group have less time spent than incongruent group.

## 5. T-test

In this test , we use 95% confidence level

```python
stats.ttest_rel(troop["Congruent"],troop["Incongruent"])
```

We can see p-value is very small , so we reject the null hypothesis and conclude that Congruent and Incongruent group are significantly different.
