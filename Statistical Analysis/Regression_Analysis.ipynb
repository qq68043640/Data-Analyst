```python
import pandas as pd
from pandas import DataFrame
import sklearn 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
BD=pd.read_csv("BankingData.csv")
```

```python
BD.head()
```

```python
lm=LinearRegression()
lm
```

```python
x=BD.drop("Balance",axis=1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,BD["Balance"],test_size=0.33,random_state=5)
```

```python
lm.fit(X_train,Y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
coef_table=pd.DataFrame(zip(x.columns,lm.coef_),columns=["feature","estimateCoefficients"])
coef_table
```

```python
plt.scatter(Y_test,lm.predict(X_test))
plt.show()
```

```python
mean_squared_error(Y_test,lm.predict(X_test))
```

```python
r2_score(Y_test,lm.predict(X_test))
```
