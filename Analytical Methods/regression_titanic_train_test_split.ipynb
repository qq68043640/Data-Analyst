```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import *
from sklearn.linear_model import *
```

```python
data=pd.read_csv("titanic-data.csv")
```

```python
data.head(20)
```

```python
def check_null():
    result={}
    for column in data:
        result[column]=data[column].isnull().values.any()
    null_result=pd.DataFrame(result,index=["IsNull"])
    return null_result
```

```python
check_null()
```

```python
age_mean=round(data["Age"].mean(),0)
data["Age"]=data["Age"].fillna(age_mean)
data["Embarked"] = data["Embarked"].fillna("S")
```

```python
Sex_map={"male":0,"female":1}
data["Sex"]=data["Sex"].map(Sex_map)
Embarked_map={"S":0,"C":1,"Q":2}
data["Embarked"]=data["Embarked"].map(Embarked_map)
```

```python
X=data.loc[:,["Pclass","Sex","Age","SibSp","Embarked","Parch","Fare"]]
Y=data.loc[:,["Survived"]]
```

```python
lm=LinearRegression()
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.33,random_state=5)
```

```python
lm.fit(X_train,Y_train)
```

```python
test_predictions=lm.predict(X_test)
```

```python
test_predictions[test_predictions>0.5]=1
test_predictions[test_predictions<0.5]=0
```

```python
accuracy=np.count_nonzero(Y_test==test_predictions)*1.0/Y_test["Survived"].count()
accuracy
```
